# load the results files

```python
import pandas as pd

dfs = {
    "linear": pd.read_csv("results_linear.csv"),
    "logistic": pd.read_csv("results_logistic.csv"),
    "mlp": pd.read_csv("results_mlp.csv"),
    #"tabnet": pd.read_csv("results_tabnet.csv")
}

results = None

for name, df in dfs.items():
    temp = df[["expected","predicted","correct"]].copy()
    temp = temp.rename(columns={"predicted":f"{name}_predicted", "correct":f"{name}_correct"})

    if results is None:
        results = temp
    else:
        temp = temp.drop(columns=["expected"])
        results = pd.concat([results, temp], axis=1)
```
